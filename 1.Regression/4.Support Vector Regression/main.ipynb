```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('tvmarketing.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-2:-1].values
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
```

```python
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_pred = sc_y.inverse_transform(y_pred.reshape(-1,1))
```

```python
y_pred
```

```python
y_test.shape
```

```python
c = [i for i in range(1,41,1)]         
fig = plt.figure()
plt.plot(c,sc_y.inverse_transform(y_test), color="blue", linewidth=2, linestyle="-")
plt.plot(c,y_pred, color="red",  linewidth=2, linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)            
plt.xlabel('Index', fontsize=18)                               
plt.ylabel('Sales', fontsize=16)  
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
r_squared = r2_score(sc_y.inverse_transform(y_test), y_pred)
mae = mean_absolute_error(sc_y.inverse_transform(y_test), y_pred)
```

```python
print('r_square_value :',r_squared)
print("Mean Absolute Error (MAE):", mae)
```

##### R-squared is 0.996, indicating that approximately 99.6% of the variance in the target variable is explained by the model.
##### MAE of 4.66 means, on average, model's predictions are off by approximately 4.66 units from the actual values.
