```python
import pandas as pd
```

## Data Preprocessing




```python
dataset = pd.read_csv('TV_Shows.csv', header = None)
```

```python
dataset.head()
```

```python
dataset.drop_duplicates(inplace = True)
```

```python
dataset
```

```python
rows,columns = dataset.shape
print(rows)
print(columns)
```

```python
transactions = []
for i in range(rows):
  transactions.append( [str(dataset.values[i,j]) for j in range(columns)])
```

```python
print(transactions)
```

## Applying Apriori Model on Dataset

```python
from apyori import apriori
result = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2,max_length = 2)
```

## Visualising Results

```python
results = list(result)
```

```python
print(results)
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame
```

```python
resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
